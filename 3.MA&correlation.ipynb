Multivariate Analysis
Scatter Plot (Discount vs Quantity Sold)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("amazon_sales_dataset.csv")

plt.scatter(df['discount_percent'], df['quantity_sold'])
plt.title("Discount vs Quantity Sold")
plt.xlabel("Discount %")
plt.ylabel("Quantity Sold")
plt.show()
```

Scatter Plot (Price vs Revenue)

```python
plt.scatter(df['price'], df['total_revenue'])
plt.title("Price vs Revenue")
plt.xlabel("Price")
plt.ylabel("Revenue")
plt.show()
```

Correlation Analysis

```python
correlation = df.corr(numeric_only=True)
print(correlation)
```

Correlation Heatmap

```python
plt.figure(figsize=(10,6))
plt.imshow(correlation)
plt.colorbar()
plt.xticks(range(len(correlation.columns)), correlation.columns, rotation=90)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.title("Correlation Heatmap")
plt.show()
```

Dashboard Ready Charts.
 
Revenue by Category

```python
df.groupby('product_category')['total_revenue'].sum().plot(kind='bar')
plt.title("Revenue by Category")
plt.show()
```

Revenue by Region

```python
df.groupby('customer_region')['total_revenue'].sum().plot(kind='bar')
plt.title("Revenue by Region")
plt.show()
```

Payment Method Share

```python
df['payment_method'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Payment Method Distribution")
plt.ylabel('')
plt.show()
```

Important KPIs

```python
print("Total Revenue:", df['total_revenue'].sum())
print("Total Orders:", df.shape[0])
print("Average Rating:", df['rating'].mean())
print("Average Discount:", df['discount_percent'].mean())
print("Average Quantity Sold:", df['quantity_sold'].mean())
```
