```python
import pandas as pd
df = pd.read_csv("amazon_sales_dataset.csv")
```

Top 5 Product Categories by Revenue

```python
top_categories = df.groupby('product_category')['total_revenue'].sum().sort_values(ascending=False).head(5)
print(top_categories)
```

Region with Highest Sales

```python
region_sales = df.groupby('customer_region')['total_revenue'].sum().sort_values(ascending=False)
print(region_sales)
```


Monthly Sales Trend

```python
import pandas as pd
import matplotlib.pyplot as plt

df['order_date'] = pd.to_datetime(df['order_date'])
df['month'] = df['order_date'].dt.to_period('M')

monthly_sales = df.groupby('month')['total_revenue'].sum()

monthly_sales.plot(kind='line')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

Most Used Payment Method

```python
payment_counts = df['payment_method'].value_counts()
print(payment_counts)
```

Top 10 Revenue Generating Products

```python
top_products = df.groupby('product_id')['total_revenue'].sum().sort_values(ascending=False).head(10)
print(top_products)
```
